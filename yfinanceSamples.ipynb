## Convert data type from object to float, so that we can plot the chart

```python
pfizerdata = pfizerdata.reset_index()
for i in ['Open', 'High', 'Close', 'Low']: 
      pfizerdata[i]  =  pfizerdata[i].astype('float64')
```

```python
pfizerdata[1:2]
```

```python
fig = go.Figure(data=[go.Candlestick(x=pfizerdata['Date'],open=pfizerdata['Open'],high=pfizerdata['High'],low=pfizerdata['Low'],close=pfizerdata['Close'])])
fig.show()
```

```python
import plotly.express as px
fig = px.line(pfizerdata, x="Date", y="Open", title='PFizer Stock Prices')
fig.show()
```

```python
fig = go.Figure(data=go.Ohlc(x=pfizerdata['Date'],open=pfizerdata['Open'],high=pfizerdata['High'],low=pfizerdata['Low'],close=pfizerdata['Close']))
fig.show()
```
